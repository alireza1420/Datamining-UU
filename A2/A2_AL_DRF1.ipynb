```python
def confidence(X, Y, patterns, num_records):
    # Support(X ∪ Y) / Support(X)
    support_XY = patterns.get(tuple(sorted(X + Y)), 0)
    support_X = patterns.get(tuple(sorted(X)), 0)
    if support_X == 0:
        return 0
    return support_XY / support_X

# Example: real -> estate
conf_real_to_estate = confidence(['real'], ['estate'], patterns, num_records)
print("Confidence real -> estate:", conf_real_to_estate)
```
